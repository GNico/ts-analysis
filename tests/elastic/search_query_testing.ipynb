```python
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch()
```

```python
index_pattern = 'movi*'
tags = ['sa_Neutro',]
context = [ 'twitter' ]
start = ''
end = ''
interval = '1H'

query = {
  "query": {
    "bool": {
      "filter": []
    }
  },
  "aggs": {
    "my_aggregation": {
      "date_histogram": {
        "field":     "@timestamp",
        "interval":  interval
      }
    }
  }
}

if start or end:
    dict = {
        "range": {
            "@timestamp": {}
        }
    }
    if start:
        dict['range']['@timestamp']['gte'] = start
    if end:
        dict['range']['@timestamp']['lte'] = end
    query['query']['bool']['filter'].append(dict)
    
if tags:
    dict = {
        "terms": {"tag.keyword": tags }
    }
    query['query']['bool']['filter'].append(dict)

if context:
    dict = {
        "terms": {"context.keyword": context }
    }
    query['query']['bool']['filter'].append(dict)
    

        
#print(query)
es.search(index=index_pattern, body=query)
```
