Para usar este script tal como esta, hay que ponerlo en el directorio de la data. Asume que en ese directorio esta la data en subdirectorios separados por clientes (el nombre de las carpetas debe ser el mismo que client_data['name'] ). El script crea una nueva carpeta "newdata" adento de cada uno con la nueva data con contextos agregados, si ya existe una carpeta con ese nombre, no hace nada.

```python
import os
import json

client_data = [
    {
        'name': 'movistar',
        'code': '31',
        'contexts': [
            'MovistarMonitFBMessage',
            'MovistarMonitFBComment',
            'MovistarMonitFBPost',
            'MVPersonalMonitFBComment',
            'MVPersonalMonitFBPost',
            'MVClaroMonitFBComment',
            'MVClaroMonitFBPost',
            'twitter',
            'instagramMedia',
            'instagramComment'
        ]
    },
    {
        'name': 'fravega',
        'code': '211',
        'contexts': [
            'email', 
            'FravegaFBComment',
            'FravegaFBMessage',
            'FravegaFBPost',
            'FravegaGarbarinoFBComment',
            'FravegaGarbarinoFBPost',
            'FravegaCompumundoFBComment',
            'FravegaCompumundoFBPost',
            'FravegaMusimundoFBComment',
            'FravegaMusimundoFBPost',
            'FravegaCetrogarFBPost',
            'FravegaCetrogarFBComment',
            'FravegaMegatoneFBComment',
            'FravegaMegatoneFBPost',
            'FravegaCastilloFBComment',
            'FravegaCastilloFBPost',
            'FravegaNews',
            'FravegaYT',
            'twitter',
            'instagramComment',
            'instagramMedia'
        ]
    },
    {
        'name': 'presidencia',
        'code': '178',
        'contexts': [
            'email'
        ]
    },
    {
        'name': 'provincia',
        'code': '295',
        'contexts': [
            'BancoProvinciaFBComment',
            'BancoProvinciaFBPost',
            'BancoProvinciaGaliciaFBPost',
            'BancoProvinciaGaliciaFBComment',
            'BancoProvinciaSantanderFBPost',
            'BancoProvinciaSantanderFBComment',
            'twitter'
        ]
    },
    {
        'name': 'despegarES',
        'code': '301',
        'contexts': [
            'DespegarViajesNPSES',
            'DespegarInterNPSES',
            'DespegarSolicitudesNPSES',
            'DespegarReproNPSES'
        ]
    },
    {
        'name': 'despegarBR',
        'code': '302',
        'contexts': [
            'DespegarViajesNPSPT',
            'DespegarInterNPSPT',
            'DespegarSolicitudesNPSPT',
            'DespegarReproNPSPT'
        ]
    }, 
]
```

```python
for client in client_data:
    destdir = './' + client['name']
    files = [ f for f in os.listdir(destdir) if os.path.isfile(os.path.join(destdir,f)) ]
    #create directory for new data
    newdirectory = destdir + '/newdata'
    if not os.path.exists(newdirectory):
        os.makedirs(newdirectory)
        for filename in files:
            if '.json' and 'contents_' + client['code'] in filename:
                with open(os.path.join(destdir,filename)) as f:
                    data = json.load(f)
                    newdata = [] 
                    for event in data:
                        matching = [c for c in client['contexts'] if c in event['_id']]
                        event['source']['context'] =  matching[0] if matching else ''
                        newdata.append(event)
                    with open(os.path.join(newdirectory,filename), "w+") as outfile:  
                        json.dump(newdata, outfile)
    else:
        print('se ignoro ' + client['name'] + 'porque el directorio ' + newdirectory + ' ya existe')

```
